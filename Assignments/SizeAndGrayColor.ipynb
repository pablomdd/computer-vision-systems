```python
from skimage import io, color, transform
import matplotlib.pyplot as plt

img = io.imread('baboon.png')
img_gray = color.rgb2gray(img)
img_hsv = color.rgb2hsv(img)
img_resized = transform.resize(img, (1024, 1024))

plt.subplot(121), io.imshow(img)
plt.title('RGB Format') 

plt.subplot(122), io.imshow(img_gray)
plt.title('Grayscale Format') 

plt.show()
```

```python
io.imsave("baboon_gray.png", img_gray)
io.imsave("baboon_hsv.png", img_hsv)
io.imsave("baboon_resized.png", img_resized)
```
